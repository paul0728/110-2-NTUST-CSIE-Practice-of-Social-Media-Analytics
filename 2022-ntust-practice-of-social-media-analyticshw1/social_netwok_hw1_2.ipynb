```python
# node_list=pd.concat([predict['node1'], predict['node2']], ignore_index=True).unique()

# node_list



G_data_all=nx.from_pandas_edgelist(pd.concat([train[['node1','node2']], predict[:500]], ignore_index=True), "node1", "node2", create_using=nx.DiGraph())


for n in pd.concat([predict['node1'], predict['node2']], ignore_index=True).unique():
    G_data_all.add_node(n)
print(nx.info(G_data_all))

# G_data_all.nodes
```

```python
from node2vec import Node2Vec

# Generate walks
node2vec= Node2Vec(G_data_all, dimensions=100, walk_length=16, num_walks=50)


# train node2vec model
n2w_model = node2vec.fit(window=7, min_count=1)
```

```python
# x = [(n2w_model[str(i)]+n2w_model[str(j)]) for i,j in zip(data['node1'], data['node2'])]
# for i,j in zip(data['node1'], data['node2']):
#     print(i,j)
xtrain = [(n2w_model.wv[list(G_data_all.nodes).index(i)]+n2w_model.wv[list(G_data_all.nodes).index(j)]) for i,j in zip(data['node1'], data['node2'])]
ytrain = data['link']
xtest = [(n2w_model.wv[list(G_data_all.nodes).index(i)]+n2w_model.wv[list(G_data_all.nodes).index(j)]) for i,j in zip(predict['node1'], predict['node2'])]
```
